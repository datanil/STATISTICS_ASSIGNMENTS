### **Q1. What is the Probability density function?**

###***ANSWER :***

**The Probability Density Function (PDF)** is a fundamental concept in probability theory and statistics. It describes the probability distribution of a continuous random variable. In simpler terms, the PDF provides the likelihood of a continuous random variable falling within a particular range of values.

***For a continuous random variable X, the PDF is denoted as f(x) and satisfies the following properties:***

1. ***f(x) is non-negative for all x.***

2. ***The total area under the curve of the PDF is equal to 1, which means the probability of the random variable taking any value within the entire range is 1.***

- Mathematically, the probability that X lies in the interval [a, b] is given by the integral of the PDF over that interval:

> > ***P(a ≤ X ≤ b) = ∫[a to b] f(x) dx***

- The PDF allows us to calculate probabilities of various events, such as the probability that X falls within a specific range or the probability of X being greater than a certain value.

- It's important to note that the PDF is applicable to continuous random variables, as opposed to discrete random variables, which are described by Probability Mass Functions (PMF). The PMF provides the probabilities for each possible discrete value that a random variable can take.

### **Q2. What are the types of Probability distribution?**
###***ANSWER :***

Probability distributions are mathematical functions that describe the likelihood of different outcomes in a probability space. There are various types of probability distributions, and they can be broadly categorized into two main types: discrete probability distributions and continuous probability distributions.

1. **Discrete Probability Distributions:**
>
Discrete probability distributions are used to model random variables that can only take on distinct, separate values. Each value has an associated probability. The probabilities for all possible values of the random variable must sum up to 1. Examples of discrete probability distributions include:

- ***Bernoulli Distribution:***
>
 Models a random variable with two possible outcomes, usually denoted as 0 and 1. It is used for binary events, like success/failure or heads/tails.
- ***Binomial Distribution:***
>
 Describes the number of successes in a fixed number of independent Bernoulli trials.
- ***Poisson Distribution:***
>
 Models the number of events that occur in a fixed interval of time or space, assuming the events occur independently and at a constant rate.
- ***Geometric Distribution:***
>
 Represents the number of Bernoulli trials needed until the first success.
- ***Hypergeometric Distribution:***
>
 Models the probability of successes in a fixed sample without replacement from a finite population.

2. **Continuous Probability Distributions:**
>
Continuous probability distributions are used to model random variables that can take on any value within a continuous range. The probability of any specific value is zero, but probabilities are defined for intervals of values. Examples of continuous probability distributions include:

- ***Uniform Distribution:***
>
 Has a constant probability density over a defined interval, resulting in a flat, rectangular-shaped distribution.
- ***Normal Distribution (Gaussian Distribution):***
>
 One of the most common distributions, characterized by a bell-shaped curve and defined by its mean and standard deviation.
- ***Exponential Distribution:***
>
 Often used to model the time between events in a Poisson process (events that occur at a constant rate independently).
- ***Beta Distribution:***
>
 Used to model probabilities when there is a limited range (between 0 and 1).
- ***Gamma Distribution:***
>
 Generalizes the exponential and chi-square distributions and is used in various applications, including queuing theory and reliability analysis.

*These are just some of the many probability distributions that exist, and each serves different purposes in modeling various real-world phenomena in different fields such as physics, engineering, economics, and social sciences.*

### **Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.**

###***ANSWER :***

The Binomial distribution has several key properties, which make it a valuable tool for modeling certain types of events. The properties of the Binomial distribution are as follows:

1. **Fixed Number of Trials:**
>
 The Binomial distribution deals with a fixed number of independent trials, denoted by 'n.' Each trial can have only two outcomes, ***usually referred to as "success" and "failure."***

2. **Independent Trials:**
>
 Each trial is assumed to be independent of the others, meaning the outcome of one trial does not affect the outcomes of the other trials.

3. **Constant Probability of Success:**
>
 The probability of success in each trial is represented by 'p,' and it remains constant across all trials.

4. **Discrete Outcome:**
>
 The random variable representing the number of successes, denoted by 'k,' can only take discrete integer values from 0 to 'n.'

5. **Probability Mass Function (PMF):**
>
 The probability of getting exactly 'k' successes in 'n' trials is given by the Binomial probability mass function:

   > ***P(X = k) = C(n, k) * p^k * (1-p)^(n-k)***

   **where C(n, k) is the binomial coefficient, which represents the number of ways to choose 'k' successes out of 'n' trials.**

***Examples of events where the Binomial distribution can be applied:***

1. **Coin Toss:**
>
 Consider flipping a fair coin multiple times. Each flip can result in either heads (success) or tails (failure). The outcome of one flip does not influence the outcome of the others, and the probability of getting heads (success) remains the same for each flip. The Binomial distribution can be used to model the number of heads obtained in 'n' coin tosses.

2. **Quality Control:**
>
 Suppose a factory produces light bulbs, and each bulb has a probability 'p' of being defective. The factory produces a fixed number 'n' of bulbs. The Binomial distribution can be used to model the number of defective bulbs in the sample of 'n' bulbs produced by the factory.

*In both examples, the Binomial distribution is suitable because there is a fixed number of independent trials, each with two possible outcomes (success or failure), and the probability of success (or getting heads in the case of coin tosses) remains constant across all trials.*

### **Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.**

###***ANSWER :***

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Parameters:
        prob_success (float): The probability of success (p) in each trial.
        sample_size (int): The size of the random sample.
"""
prob_success = 0.4
sample_size = 1000
binomial_dist = np.random.binomial(1 , prob_success , sample_size)

plt.hist(binomial_dist)
plt.xticks([0,1] , ['Failure' , 'Success'])
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title(f"Histogram of {sample_size} Random Samples from a Binomial Distribution (p={prob_success})")
plt.show()
```

### **Q7. How Binomial distribution different from Poisson distribution?**

###***ANSWER :***

The Binomial distribution and the Poisson distribution are both probability distributions used to model discrete random variables. However, they have distinct characteristics and are applied in different situations.

The key difference between the Binomial and Poisson distributions lies in the context and assumptions of their applications. The Binomial distribution is used for a fixed number of independent trials with two outcomes, while the Poisson distribution is used for modeling the occurrence of rare events in a fixed interval with a constant average rate of occurrence.

1. **Binomial Distribution:**
>
The Binomial distribution is used to model the number of successes in a fixed number of independent Bernoulli trials. In each trial, there are only two possible outcomes, usually denoted as "success" and "failure." The trials are assumed to be independent and have a constant probability of success, denoted by 'p,' throughout all the trials.

***Properties of the Binomial distribution:***
- The random variable represents the number of successes (k) in 'n' trials.

- The probability mass function (PMF) of the Binomial distribution is given by: P(X = k) = C(n, k) * p^k * (1-p)^(n-k), where C(n, k) is the binomial coefficient or the number of ways to choose 'k' successes from 'n' trials.
- The Binomial distribution is discrete and defined for discrete integer values of 'k' (0, 1, 2, ..., n).

The Binomial distribution is commonly used when dealing with situations where there are only two possible outcomes in each trial and the trials are independent, such as coin tosses, success/failure experiments, or the number of defective items in a sample.

2. **Poisson Distribution:**
>
The Poisson distribution is used to model the number of events that occur in a fixed interval of time or space, given that the events occur independently and at a constant average rate (mean), denoted by 'λ' (lambda). It is often used to model rare events where the average rate of occurrence is small.

***Properties of the Poisson distribution:***

- The random variable represents the number of events (k) occurring in a fixed interval.

- The probability mass function (PMF) of the Poisson distribution is given by: P(X = k) = (e^(-λ) * λ^k) / k!, where 'e' is Euler's number (approximately 2.71828).

- The Poisson distribution is discrete and defined for discrete integer values of 'k' (0, 1, 2, ...).

*The Poisson distribution is commonly used in situations like modeling the number of customer arrivals at a service center, the number of accidents in a day, or the number of emails received per hour.*



### **Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.**

###***ANSWER :***

```python
import numpy as np

# Parameters
mean = 5
sample_size = 1000

sample = np.random.poisson(mean, sample_size)
# Calculate the sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample, ddof=1)


print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

### **Q9. How mean and variance are related in Binomial distribution and Poisson distribution?**

###***ANSWER :***

In both the Binomial distribution and the Poisson distribution, the mean and variance are related to the parameters of the respective distributions.
For both the Binomial and Poisson distributions, the mean represents the center or average value, while the variance measures the spread or variability of the outcomes around the mean. In the Binomial distribution, the variance depends on both the number of trials (n) and the probability of success (p), while in the Poisson distribution, the variance is solely determined by the average rate (λ).
 Let's explore this relationship for each distribution:

1. **Binomial Distribution:**
>
In the Binomial distribution, the mean (μ) and variance (σ^2) are related as follows:

>***Mean (μ) = n * p***

> ***Variance (σ^2) = n * p * (1 - p)***

**where:**
- **n is the number of trials in the Binomial distribution.**
- **p is the probability of success in each trial.**

The mean represents the average number of successes in 'n' trials, while the variance represents the spread or dispersion of the number of successes around the mean. As 'p' gets closer to 0.5, the variance reaches its maximum value of n/4, indicating a higher spread of outcomes around the mean.

2. **Poisson Distribution:**
>
In the Poisson distribution, the mean (λ) and variance (σ^2) are related as follows:

> ***Mean (λ) = Variance (σ^2) = λ***

**where:**
- **λ (lambda) is the average rate of events in a fixed interval of time or space.**

In the Poisson distribution, the mean and variance are both equal to λ, indicating that the average rate of events is also the measure of spread or dispersion around the mean. The Poisson distribution is often used to model rare events, and when the average rate (λ) is large, the variance also increases, indicating a higher spread of events around the mean.

In summary, for both the Binomial and Poisson distributions, the mean represents the center or average value, while the variance measures the spread or variability of the outcomes around the mean. In the Binomial distribution, the variance depends on both the number of trials (n) and the probability of success (p), while in the Poisson distribution, the variance is solely determined by the average rate (λ).

### **Q10. In normal distribution with respect to mean position, where does the least frequent data appear?**

###***ANSWER :***

In a normal distribution, the least frequent data appears in the tails of the distribution, farthest from the mean. The normal distribution is a symmetric bell-shaped curve, and the data becomes less frequent as it moves away from the mean in either direction towards the tails.

The probability density is highest at the mean, which represents the most common or typical value in the distribution. As you move away from the mean towards the tails, the probability density decreases, indicating that the values in the tails are less likely to occur.

In a standard normal distribution (mean = 0 and standard deviation = 1), approximately 68% of the data lies within one standard deviation of the mean (within the range [mean - 1 * standard deviation, mean + 1 * standard deviation]). As you move two standard deviations away from the mean, approximately 95% of the data lies within this range ([mean - 2 * standard deviation, mean + 2 * standard deviation]). And about 99.7% of the data lies within three standard deviations from the mean ([mean - 3 * standard deviation, mean + 3 * standard deviation]).

This means that the data in the tails, which are further than three standard deviations away from the mean, are the least frequent data points in a normal distribution. These extreme values in the tails are considered outliers or rare occurrences. The majority of the data concentrates around the mean and becomes less frequent as you move away from it towards the tails.
