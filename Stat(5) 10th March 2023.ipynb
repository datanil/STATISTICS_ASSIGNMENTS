### **Q1: What is Estimation Statistics? Explain point estimate and interval estimate.**

### ***ANSWER :***

Estimation statistics is a branch of statistics that deals with estimating population parameters based on sample data. When it is impractical or impossible to collect data from an entire population, statisticians use samples to make inferences about the population. Estimation involves using sample statistics to estimate unknown population parameters.

There are two main types of estimation: point estimate and interval estimate.

1. **Point Estimate:**
>
A point estimate is a single numerical value that serves as the best guess or estimate of the unknown population parameter. It is obtained by using a statistic computed from the sample data. For example, if we want to estimate the population mean (μ) based on a sample of data, the sample mean (x̄) is used as the point estimate for μ. Similarly, if we want to estimate the population proportion (p) based on a sample, the sample proportion (p̂) is used as the point estimate for p.

The point estimate provides a single value but does not give any information about the precision or accuracy of the estimate. To address this limitation, interval estimation is used.

2. **Interval Estimate:**
>
Interval estimation provides a range of values within which the population parameter is likely to lie, along with a level of confidence associated with this range. This range is called a confidence interval. The confidence interval is constructed around the point estimate, and it quantifies the uncertainty of the estimate.

*For example, a 95% confidence interval for the population mean (μ) would be written as (x̄ - E, x̄ + E), where x̄ is the sample mean and E is the margin of error. The margin of error depends on the variability of the data and the desired level of confidence. The interpretation of a 95% confidence interval is that if we were to repeat the sampling process many times and construct confidence intervals for each sample, approximately 95% of these intervals would contain the true population mean.*



### **Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.**

### ***ANSWER :***

```python
import numpy as np
import scipy.stats as stats

# Example usage:
sample_mean = 50.0
sample_std_dev = 10.0
sample_size = 100
confidence_level = 0.95

# Calculate the standard error of the mean (SE) using the sample standard deviation and sample size
standard_error = sample_std_dev / np.sqrt(sample_size)

# Calculate the degrees of freedom (df) for the t-distribution
degrees_of_freedom = sample_size - 1

# Calculate the critical value for the given confidence level and degrees of freedom
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha / 2, df=degrees_of_freedom)

# Calculate the margin of error (E) for the confidence interval
margin_of_error = t_critical * standard_error

# Calculate the lower and upper bounds of the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

estimated_mean , confidence_interval = sample_mean , (lower_bound, upper_bound)

print("Estimated population mean:", estimated_mean)
print(f"{confidence_level*100}% Confidence Interval: {confidence_interval}")
```

### **Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.**

### ***ANSWER :***

Hypothesis testing is a statistical method used to make decisions and draw conclusions about a population based on sample data. It involves formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (Ha), and then using sample evidence to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

Here's how the process generally works:

1. **Null Hypothesis (H0):** The null hypothesis is a statement of no effect or no difference in the population. It represents the status quo or the conventional wisdom. For example, it could state that there is no difference between two groups, no relationship between variables, or no effect of a treatment.

2. **Alternative Hypothesis (H1):** The alternative hypothesis is the statement that contradicts the null hypothesis. It represents what the researcher is trying to find evidence for. It could claim a significant difference, relationship, or effect.

3. **Test Statistic:** A test statistic is a value calculated from the sample data that helps us assess whether the null hypothesis should be rejected or not.

4. **Significance Level (α):** The significance level is the probability of making a Type I error, which is the probability of rejecting the null hypothesis when it is actually true. Commonly used significance levels are 0.05 (5%) and 0.01 (1%).

5. **P-value:** The p-value is the probability of observing the test statistic or a more extreme value, assuming that the null hypothesis is true. A small p-value (usually less than the chosen significance level) indicates strong evidence against the null hypothesis and supports the alternative hypothesis.

6. **Conclusion:** If the p-value is less than the significance level (α), we reject the null hypothesis in favor of the alternative hypothesis. If the p-value is greater than α, we fail to reject the null hypothesis, meaning we do not have sufficient evidence to support the alternative hypothesis.

***Importance of Hypothesis Testing:***
1. **Objectivity:** Hypothesis testing provides an objective and systematic approach to making decisions based on data rather than personal opinions or biases.

2. **Inference:** It allows us to make inferences about the population using sample data, even when it is not feasible to collect data from the entire population.

3. **Scientific Method:** Hypothesis testing is an essential part of the scientific method. It helps researchers formulate and test specific hypotheses, leading to advancements in knowledge and understanding of various phenomena.

4. **Decision-Making:** Hypothesis testing helps in making informed decisions in various fields, such as medicine, social sciences, business, and engineering, by providing evidence to support or reject certain claims.

5. **Control of Type I Error:** By using a predetermined significance level (α), hypothesis testing controls the risk of making a Type I error, ensuring that conclusions are valid within the given probability threshold.

6. **Statistical Comparisons:** Hypothesis testing enables researchers to compare different groups, treatments, or conditions to identify meaningful differences or effects.

- Sample Standard Deviation (s_male): 10 pounds

Sample of Female College Students:

- Sample Size (n_female): 45
- Sample Mean Weight (x̄_female): 160 pounds
- Sample Standard Deviation (s_female): 8 pounds

Now, we will perform the hypothesis test to determine whether the average weight of male college students is greater than the average weight of female college students.

Hypothesis:

- Null Hypothesis (H0): The average weight of male college students is equal to or less than the average weight of female college students.

- Alternative Hypothesis (H1): The average weight of male college students is greater than the average weight of female college students.

We can use a two-sample t-test to perform this hypothesis test. Let's assume a significance level (α) of 0.05.




```python

import scipy.stats as stats

# Example usage:
sample_mean_male = 175
sample_std_dev_male = 10
sample_size_male = 50

sample_mean_female = 160
sample_std_dev_female = 8
sample_size_female = 45

alpha = 0.05

# Calculate the standard error of the means for both samples
standard_error_male = sample_std_dev_male / (sample_size_male ** 0.5)
standard_error_female = sample_std_dev_female / (sample_size_female ** 0.5)

# Calculate the degrees of freedom for the t-distribution
degrees_of_freedom = sample_size_male + sample_size_female - 2

# Calculate the t-statistic
t_statistic = (sample_mean_male - sample_mean_female) / ((standard_error_male ** 2 / sample_size_male) + (standard_error_female ** 2 / sample_size_female)) ** 0.5

# Calculate the critical value for the given significance level and degrees of freedom
critical_value = stats.t.ppf(1 - alpha, df=degrees_of_freedom)

# Compare the t-statistic with the critical value to make a decision
print(f"T - Statistics : {t_statistic}")
print(f"Critical_value: {critical_value}")
if t_statistic > critical_value:
  print("Reject H0. There is enough evidence to support Ha. Male college students have a greater average weight than female college students.")
else:
  print("Fail to reject H0. There is not enough evidence to support Ha. We cannot conclude that male college students have a greater average weight than female college students.")

```

### **Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.**

```python
import numpy as np
import scipy.stats as stats

# Example Data:
sample1 = [28, 35, 42, 30, 36, 40, 31, 38, 32, 34]
sample2 = [33, 39, 37, 41, 36, 38, 40, 43, 45, 44]

alpha = 0.05

# Calculate the sample statistics for both samples
sample_mean1 = np.mean(sample1)
sample_std_dev1 = np.std(sample1, ddof=1)  # ddof=1 for unbiased estimation of sample variance
sample_size1 = len(sample1)

sample_mean2 = np.mean(sample2)
sample_std_dev2 = np.std(sample2, ddof=1)
sample_size2 = len(sample2)

# Calculate the degrees of freedom for the t-distribution
degrees_of_freedom = sample_size1 + sample_size2 - 2

# Calculate the pooled standard deviation
pooled_std_dev = np.sqrt(((sample_std_dev1 ** 2) + (sample_std_dev2 ** 2)) / 2)

# Calculate the t-statistic
t_statistic = (sample_mean1 - sample_mean2) / (pooled_std_dev * np.sqrt(1/sample_size1 + 1/sample_size2))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=degrees_of_freedom))


print("Results of Two-Sample t-Test:")
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_val}")

# Check if the null hypothesis is rejected or not based on the p-value and alpha
if p_val < alpha:
    print("Null hypothesis rejected. There is enough evidence to suggest that the means of the two populations are different.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest a significant difference between the means of the two populations.")
```

### **Q6: What is a null and alternative hypothesis? Give some examples.**

### ***ANSWER :***

In statistical hypothesis testing, the null hypothesis (H0) and alternative hypothesis (Ha) are two competing statements or claims about a population parameter. These hypotheses are used to make decisions based on sample data and statistical analysis. The null hypothesis represents the status quo or no effect, while the alternative hypothesis represents what the researcher wants to find evidence for.

1. Null Hypothesis (H0):

The null hypothesis is a statement that assumes there is no significant effect, no difference, or no relationship between variables in the population. It typically represents the default assumption or the absence of an effect. In hypothesis testing, we assume the null hypothesis is true until there is enough evidence to reject it in favor of the alternative hypothesis. The null hypothesis is denoted as H0.

Example 1:
H0: The mean height of adult males is equal to 70 inches.

Example 2:
H0: There is no difference in test scores between two teaching methods.

Example 3:

2. Alternative Hypothesis (H1):

The alternative hypothesis is a statement that contradicts the null hypothesis and represents what the researcher wants to find evidence for. It typically claims a significant effect, difference, or relationship in the population. In hypothesis testing, we seek evidence to support the alternative hypothesis if the data suggests a significant departure from the null hypothesis. The alternative hypothesis is denoted as Ha.

Example 1:
H1: The mean height of adult males is greater than 70 inches.

Example 2:
H1: There is a difference in test scores between two teaching methods.

Example 3:

When conducting a hypothesis test, the goal is to use sample data to assess the strength of evidence against the null hypothesis. The test provides a p-value, which represents the probability of observing the data (or more extreme) if the null hypothesis is true. If the p-value is below a chosen significance level (alpha), typically 0.05, we reject the null hypothesis in favor of the alternative hypothesis, indicating that there is enough evidence to support the claim represented by the alternative hypothesis. Otherwise, if the p-value is greater than or equal to the significance level, we fail to reject the null hypothesis due to insufficient evidence, and no definitive conclusions are drawn.

### **Q7: Write down the steps involved in hypothesis testing.**

### ***ANSWER :***

Hypothesis testing is a structured process used to make decisions about a population based on sample data. The steps involved in hypothesis testing are as follows:

1. State the Hypotheses:
Formulate the null hypothesis (H0) and the alternative hypothesis (Ha) that represent the two competing claims about the population parameter being tested. The null hypothesis typically assumes no effect or no difference, while the alternative hypothesis contradicts the null hypothesis and represents the effect or difference of interest.

2. Set the Significance Level (α):
Choose a significance level (alpha) that represents the maximum probability of making a Type I error (rejecting the null hypothesis when it is true). Commonly used values for alpha are 0.05 (5%) and 0.01 (1%).

3. Collect Sample Data:
Gather data from a random sample that is representative of the population of interest. The sample data will be used to perform the statistical test and make inferences about the population parameter.

4. Select the Appropriate Test:
Choose the appropriate statistical test based on the type of data and the research question. For example, common tests include the t-test for comparing means, chi-square test for comparing proportions, and ANOVA for comparing means of multiple groups.

5. Calculate the Test Statistic:
Compute the test statistic using the sample data and the chosen statistical test. The test statistic quantifies how far the sample estimate deviates from the null hypothesis.

6. Determine the Critical Region or P-value:
Depending on the chosen significance level (alpha) and the test statistic's distribution, identify the critical region or calculate the p-value. The critical region is the range of values that, if the test statistic falls within, leads to the rejection of the null hypothesis. The p-value represents the probability of observing the data (or more extreme) under the assumption that the null hypothesis is true.

7. Make a Decision:
Compare the test statistic with the critical value (for critical region approach) or compare the p-value with the significance level (for p-value approach). If the test statistic falls in the critical region or if the p-value is less than the significance level (alpha), reject the null hypothesis in favor of the alternative hypothesis. Otherwise, if the test statistic falls outside the critical region or the p-value is greater than or equal to alpha, fail to reject the null hypothesis.

8. Draw Conclusions:
Based on the decision made in Step 7, draw conclusions about the population parameter. If the null hypothesis is rejected, support the claim represented by the alternative hypothesis. If the null hypothesis is not rejected, there is not enough evidence to support the claim represented by the alternative hypothesis.

9. Interpretation:
Interpret the results of the hypothesis test in the context of the research question and the real-world implications. Provide a meaningful explanation of the findings and their significance.

Remember that hypothesis testing provides evidence, but it does not prove or disprove a hypothesis with certainty. It helps researchers make data-driven decisions and draw inferences about populations based on sample data.

### **Q8. Define p-value and explain its significance in hypothesis testing.**

### ***ANSWER :***

The p-value, short for "probability value," is a crucial concept in hypothesis testing. It represents the probability of obtaining the observed results (or results more extreme) when the null hypothesis is true. In other words, the p-value measures the strength of evidence against the null hypothesis based on the sample data. A smaller p-value indicates stronger evidence against the null hypothesis, while a larger p-value suggests weaker evidence.

In hypothesis testing, the significance of the p-value lies in its role in decision-making. Here's how it works:

1. **Null Hypothesis (H0) Assumption:**
>
The null hypothesis assumes that there is no significant effect, difference, or relationship between variables in the population. It is the status quo or the default assumption.

2. **Alternative Hypothesis (H1):**
>
The alternative hypothesis contradicts the null hypothesis and represents what the researcher wants to find evidence for. It claims a significant effect, difference, or relationship in the population.

3. **Decision Rule:**
>
To conduct a hypothesis test, a significance level (alpha) is chosen as the threshold for decision-making. Commonly used alpha values are 0.05 (5%) or 0.01 (1%). The significance level represents the maximum acceptable probability of making a Type I error (incorrectly rejecting the null hypothesis when it is true).

4. **Compare the p-value with Alpha:**
>
After calculating the test statistic, we obtain the p-value. If the p-value is less than or equal to the chosen significance level (p ≤ alpha), we say that the results are statistically significant at that level. In this case, we reject the null hypothesis in favor of the alternative hypothesis. This means there is enough evidence to suggest that the population parameter is likely different from the value assumed by the null hypothesis.

5. **Interpretation:**
>
If the p-value is greater than the significance level (p > alpha), we say that the results are not statistically significant at that level. In this case, we fail to reject the null hypothesis. This means there is not enough evidence to support the alternative hypothesis, and we cannot conclude that the population parameter differs significantly from the null hypothesis assumption.

***Key Points:***
- A small p-value (typically less than the chosen significance level) suggests strong evidence against the null hypothesis.
- A large p-value indicates weak evidence against the null hypothesis and suggests that the observed results are likely to occur under the assumption of the null hypothesis.
- The p-value does not give the probability that the null hypothesis is true or false. It only quantifies the evidence against the null hypothesis based on the data at hand.
- The choice of the significance level (alpha) is important. A lower alpha leads to a more stringent test and requires stronger evidence to reject the null hypothesis.

### **Q11: What is Student’s t distribution? When to use the t-Distribution.**

### ***ANSWER :***

Student's t-distribution is a probability distribution that is used in statistical inference when the sample size is small or when the population standard deviation is unknown. It is similar to the standard normal distribution (Z-distribution) but has heavier tails, which makes it more suitable for smaller sample sizes.

The t-distribution is named after William Sealy Gosset, who used the pseudonym "Student" when publishing his work due to the confidentiality of his employer. Gosset developed the t-distribution to estimate the mean of a normally distributed population when the sample size is small.

***Characteristics of the t-distribution:***
- The t-distribution is bell-shaped and symmetric like the standard normal distribution.
- The shape of the t-distribution is determined by the degrees of freedom (df). The degrees of freedom depend on the sample size and define the number of values that are free to vary in the calculation of the t-statistic.
- As the sample size (degrees of freedom) increases, the t-distribution becomes closer to the standard normal distribution.

***When to use the t-Distribution:***
The t-distribution is used in several scenarios:

1. **Small Sample Size:** When the sample size is small (typically less than 30) and the population standard deviation is unknown, the t-distribution should be used instead of the standard normal distribution for hypothesis testing and confidence interval calculations.

2. **Population Standard Deviation Unknown:** If the population standard deviation is unknown, and the sample size is small or large, the t-distribution is more appropriate for estimating the population mean.

3. **Comparing Two Sample Means:** When comparing the means of two independent samples (two-sample t-test) or paired samples (paired t-test), the t-distribution is used to assess whether the means are significantly different.

4. **Confidence Intervals(CI):** For constructing confidence intervals for population parameters, such as the population mean, when the sample size is small or the population standard deviation is unknown.


### **Q12: What is t-statistic? State the formula for t-statistic.**

### ***ANSWER :***

The t-statistic is a measure used in hypothesis testing to assess whether the sample data provides enough evidence to reject the null hypothesis. It compares the difference between the sample mean and the hypothesized population mean to the variability (standard error) of the sample mean. The t-statistic quantifies how many standard errors the sample mean is away from the hypothesized population mean under the assumption that the null hypothesis is true.

The formula for the t-statistic depends on the type of hypothesis test being performed. Here are the formulas for two common cases:

1. **One-sample t-test (testing a single sample mean against a known or hypothesized population mean):**
   The formula for the t-statistic is given by:
   
   **t = (x̄ - μ) / (s / √n)**

   **where:**
   - **t is the t-statistic,**
   - **x̄ is the sample mean,**
   - **μ is the hypothesized population mean under the null hypothesis,**
   - **s is the sample standard deviation,**
   - **n is the sample size.**

2. **Two-sample t-test (comparing two independent sample means):**
   The formula for the t-statistic is given by:
   
   **t = (x̄1 - x̄2) / √((s1^2 / n1) + (s2^2 / n2))**

   **where:**
   - **t is the t-statistic,**
   - **x̄1 and x̄2 are the sample means of the two independent samples,**
   - **s1 and s2 are the sample standard deviations of the two samples,**
   - **n1 and n2 are the sample sizes of the two samples.**

The t-statistic is used to calculate the p-value, which represents the probability of observing the data (or more extreme data) under the assumption that the null hypothesis is true. A smaller p-value indicates stronger evidence against the null hypothesis, supporting the alternative hypothesis. A larger p-value suggests weaker evidence, indicating that the data is consistent with the null hypothesis.

The first group (n1 = 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.**

### ***ANSWER :***

```python
import scipy.stats as stats

# Given data for Group 1
sample_mean1 = 80
sample_std_dev1 = 10
sample_size1 = 30

# Given data for Group 2
sample_mean2 = 75
sample_std_dev2 = 8
sample_size2 = 40

alpha = 0.01
# Calculate the degrees of freedom for the t-distribution
degrees_of_freedom = sample_size1 + sample_size2 - 2

# Calculate the pooled standard deviation (for unequal sample sizes)
pooled_std_dev = ((sample_size1 - 1) * sample_std_dev1 ** 2 + (sample_size2 - 1) * sample_std_dev2 ** 2) / degrees_of_freedom
pooled_std_dev = pooled_std_dev ** 0.5

# Calculate the standard error of the difference between means
standard_error = pooled_std_dev * ((1 / sample_size1) + (1 / sample_size2)) ** 0.5

# Calculate the t-statistic
t_statistic = (sample_mean1 - sample_mean2) / standard_error

# Calculate the two-tailed p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_of_freedom))


# Print the result
print("Results of Two-Sample t-Test:")
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_val}")

# Check if the null hypothesis is rejected or not based on the p-value and alpha
if p_val < alpha:
    print("Null hypothesis rejected. There is enough evidence to suggest that the population means for the two groups are not equal.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest a significant difference in population means for the two groups.")
```

### **Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.**

### ***ANSWER :***

```python
import scipy.stats as stats

# Given data
sample_mean = 4
sample_std_dev = 1.5
sample_size = 50
confidence_level = 0.99

# Calculate the critical value for the t-distribution
degrees_of_freedom = sample_size - 1
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha / 2, df=degrees_of_freedom)

# Calculate the standard error of the mean
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error
margin_of_error = t_critical * standard_error

# Calculate the lower and upper bounds of the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the result
print(f"Population Mean Number of Ads Watched (99% Confidence Interval): {lower_bound:.2f} to {upper_bound:.2f}")
```
