### **Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.**

```python
import scipy.stats as stats

# Given data
sample_mean = 50
standard_deviation = 5
sample_size = 100  #  the sample size should be provided here

# Calculate the critical value (t_critical) for 95% confidence level
confidence_level = 0.95
degrees_of_freedom = sample_size - 1
t_critical = stats.t.ppf(confidence_level, degrees_of_freedom)

# Calculate the standard error
standard_error = standard_deviation / (sample_size ** 0.5)

# Calculate the confidence interval
lower_bound = sample_mean - t_critical * standard_error
upper_bound = sample_mean + t_critical * standard_error

# Print the results
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

### **Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.**

```python
import numpy as np
import pandas as pd
from scipy.stats import chisquare

# Observed frequencies of colors in the bag (provide actual counts)
observed_frequencies = np.array([20, 15, 25, 10, 10, 20])

# Expected frequencies based on the theoretical distribution
expected_frequencies = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed_frequencies)

# Create a DataFrame for observed and expected frequencies
data = pd.DataFrame({
    'Observed': observed_frequencies,
    'Expected': expected_frequencies
}, index=['Blue', 'Orange', 'Green', 'Yellow', 'Red', 'Brown'])

# Perform the chi-square goodness of fit test
chi2, p_value = chisquare(f_obs=data['Observed'], f_exp=data['Expected'])

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level (usually set at 0.05)

if p_value < alpha:
    print("The p-value is less than the significance level.")
    print("There is evidence to reject the null hypothesis.")
    print("The observed distribution of colors does not match the expected distribution.")
else:
    print("The p-value is greater than the significance level.")
    print("There is no evidence to reject the null hypothesis.")
    print("The observed distribution of colors matches the expected distribution.")
```

### **Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.**

```python
import scipy.stats as stats
import math

# Given data
sample_size = 500
sample_proportion = 60 / sample_size

# Calculate the critical value (Z) for 95% confidence level
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the standard error
standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the confidence interval
lower_bound = sample_proportion - z_critical * standard_error
upper_bound = sample_proportion + z_critical * standard_error

# Print the results
print("95% Confidence Interval for the proportion of smokers: ({:.4f}, {:.4f})".format(lower_bound, upper_bound))
```

### **Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the degrees of freedom
df = 10

# Create a range of x values for the plot
x = np.linspace(0, 30, 1000)

# Calculate the probability density function (PDF) for the chi-square distribution
pdf = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, pdf, label=f"Chi-Square with {df} degrees of freedom")
plt.xlabel("Chi-Square Statistic")
plt.ylabel("Probability Density Function (PDF)")
plt.title(f"Chi-Square Distribution with {df} degrees of freedom")

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
pdf_fill = stats.chi2.pdf(x_fill, df)
plt.fill_between(x_fill, pdf_fill, color='gray', alpha=0.5, label="Area for chi-square = 15")

plt.legend()
plt.grid(True)
plt.show()
```

### **Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.**

```python
import scipy.stats as stats
import math

# Given data
sample_size = 1000
sample_proportion = 520 / sample_size

# Calculate the critical value (Z) for 99% confidence level
confidence_level = 0.99
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the standard error
standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the confidence interval
lower_bound = sample_proportion - z_critical * standard_error
upper_bound = sample_proportion + z_critical * standard_error

# Print the results
print("99% Confidence Interval for the proportion who prefer Coke: ({:.4f}, {:.4f})".format(lower_bound, upper_bound))
```

### **Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.**

```python
import numpy as np
from scipy.stats import chisquare

# Given data
observed_tails = 45
total_flips = 100

# Calculate the observed heads
observed_heads = total_flips - observed_tails

# Expected frequencies for a fair coin
expected_tails = total_flips / 2
expected_heads = total_flips / 2

# Create the observed and expected frequency arrays
observed_frequencies = np.array([observed_tails, observed_heads])
expected_frequencies = np.array([expected_tails, expected_heads])

# Perform the chi-square goodness of fit test
chi2, p_value = chisquare(observed_frequencies, expected_frequencies)

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level (usually set at 0.05)

if p_value < alpha:
    print("The p-value is less than the significance level.")
    print("There is evidence to reject the null hypothesis.")
    print("The observed frequencies do not match the expected frequencies of a fair coin.")
else:
    print("The p-value is greater than the significance level.")
    print("There is no evidence to reject the null hypothesis.")
    print("The observed frequencies match the expected frequencies of a fair coin.")
```

### **Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.**

```python
import scipy.stats as stats

# Given data
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean_hypothesis = 70
alpha = 0.05

# Calculate the standard error of the mean
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean_hypothesis) / standard_error

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the p-value
p_value = stats.t.sf(abs(t_statistic), df=degrees_of_freedom) * 2

# Print the results
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Interpret the results
if p_value < alpha:
    print("The p-value is less than the significance level.")
    print("There is evidence to reject the null hypothesis.")
    print("The population mean is significantly different from 70.")

else:
    print("The p-value is greater than the significance level.")
    print("There is no evidence to reject the null hypothesis.")
    print("The population mean is not significantly different from 70.")
```
