### **Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio test. The function should return the F-value and the corresponding p-value for the test.**

### ***ANSWER :***

```python
import numpy as np
from scipy.stats import f_oneway

# Example data:
data1 = [25, 30, 35, 40, 45]
data2 = [20, 22, 25, 28, 30]
# Convert input arrays to numpy arrays to ensure consistency
array1 = np.asarray(data1)
array2 = np.asarray(data2)

# Perform the variance ratio test using scipy's f_oneway
F, p = f_oneway(array1, array2)


F_value, p_value = F, p
print("F-value:", F_value)
print("p-value:", p_value)
```

### **Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an F-distribution, write a Python function that returns the critical F-value for a two-tailed test.**

### ***ANSWER :***

```python
from scipy.stats import f

# Example data:
significance_level = 0.05
df_num = 3
df_denom = 20

# Calculate the critical F-value using the percent point function (ppf) of the F-distribution
alpha = significance_level / 2  # Divide by 2 for a two-tailed test (splitting alpha between two tails)
critical_f = f.ppf(1 - alpha, df_num, df_denom)


critical_f_value = critical_f
print("Critical F-value:", critical_f_value)
```

### **Q3. Write a Python program that generates random samples from two normal distributions with known variances and uses an F-test to determine if the variances are equal. The program should output the F- value, degrees of freedom, and p-value for the test.**

### ***ANSWER :***

```python
import numpy as np
from scipy.stats import f_oneway

# Example usage:
mean1 = 10
std1 = 2
size1 = 30
mean2 = 12
std2 = 3
size2 = 25
# Generate random samples from the normal distributions
samples1 = np.random.normal(loc=mean1, scale=std1, size=size1)
samples2 = np.random.normal(loc=mean2, scale=std2, size=size2)


# Calculate the F-value and p-value using scipy's f_oneway
F, p = f_oneway(samples1, samples2)

# Calculate degrees of freedom for the numerator and denominator
df_num = len(samples1) - 1
df_denom = len(samples2) - 1





F_value, p_value,  = F , p

print("F-value:", F_value)
print("p-value:", p_value)
print("Degrees of freedom (numerator):", df_num)
print("Degrees of freedom (denominator):", df_denom)
```

### **Q4.The variances of two populations are known to be 10 and 15. A sample of 12 observations is taken from each population. Conduct an F-test at the 5% significance level to determine if the variances are significantly different.**

### ***ANSWER :***

```python
from scipy.stats import f

# Given data
variance1 = 10
variance2 = 15
sample_size1 = 12
sample_size2 = 12
significance_level = 0.05

# Conduct the F-test
# Calculate the F-statistic
F_statistic = variance2 / variance1

# Calculate degrees of freedom for the numerator and denominator
df_num = sample_size2 - 1
df_denom = sample_size1 - 1

# Calculate the critical F-value
critical_F = f.ppf(1 - significance_level / 2, df_num, df_denom)


# Compare F-statistic with the critical F-value and return the result
# Output the result
if F_statistic > critical_F:
    print("The variances are significantly different.")
else:
    print("The variances are not significantly different.")
```

### **Q5. A manufacturer claims that the variance of the diameter of a certain product is 0.005. A sample of 25 products is taken, and the sample variance is found to be 0.006. Conduct an F-test at the 1% significance level to determine if the claim is justified.**

### ***ANSWER :***

```python
from scipy.stats import f

# Given data
claimed_variance = 0.005
sample_variance = 0.006
sample_size = 25
significance_level = 0.01

# Calculate the F-statistic
F_statistic = sample_variance / claimed_variance

# Calculate degrees of freedom for the numerator and denominator
df_num = sample_size - 1
df_denom = 1

# Calculate the critical F-value
critical_F = f.ppf(1 - significance_level / 2, df_num, df_denom)

# Output the result
if F_statistic > critical_F or F_statistic < 1 / critical_F:
    print("The claim is not justified.")
else:
    print("The claim is justified.")
```

### **Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an F-distribution and calculates the mean and variance of the distribution. The function should return the mean and variance as a tuple.**

### ***ANSWER :***

```python
from scipy.stats import f

# Example data:
df_num = 3
df_denom = 20

# Create an instance of the F-distribution with given degrees of freedom
f_dist = f(df_num, df_denom)

# Get the mean and variance of the F-distribution
mean = f_dist.mean()
variance = f_dist.var()

print("Mean of F-distribution:", mean)
print("Variance of F-distribution:", variance)
```

### **Q7. A random sample of 10 measurements is taken from a normal population with unknown variance. The sample variance is found to be 25. Another random sample of 15 measurements is taken from another normal population with unknown variance, and the sample variance is found to be 20. Conduct an F-test at the 10% significance level to determine if the variances are significantly different.**

### ***ANSWER :***

```python
from scipy.stats import f

# Given data
sample_variance1 = 25
sample_variance2 = 20
sample_size1 = 10
sample_size2 = 15
significance_level = 0.10

# Calculate the F-statistic
F_statistic = sample_variance2 / sample_variance1

# Calculate degrees of freedom for the numerator and denominator
df_num = sample_size2 - 1 #numerator
df_denom = sample_size1 - 1 #denomenetor

# Calculate the critical F-value
critical_F = f.ppf(1 - significance_level / 2, df_num, df_denom)

# Output the result
if F_statistic > critical_F or F_statistic < 1 / critical_F:
    print("The variances are significantly different.")
else:
    print("The variances are not significantly different.")
```
